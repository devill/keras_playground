```python
from bs4 import BeautifulSoup
```

```python
def get_speech_from(filename):
    with open(filename, encoding='utf8') as f:
        data = f.read();
    soup = BeautifulSoup(data, 'html.parser')
    paragraphs = soup.find('div', {'class': 'article-content'}).find_all('p')
    return '\n\n\n' + ' '*256 + '\n' + '\n\n'.join(list(map(lambda x: x.text, paragraphs[1:-1])))
    
```

```python
from os import listdir
from os.path import isfile, join

mypath='/data/orban/speeches-fixed'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
for filename in onlyfiles:
    print(filename)
    try:
        speech = get_speech_from('/data/orban/speeches-fixed/'+filename)
        with open('/data/orban/speeches-text/'+filename,'w', encoding='utf8') as f:
            f.write(speech)
    except:
        pass
```
